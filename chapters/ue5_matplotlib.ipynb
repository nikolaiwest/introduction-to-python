#### **2.2 Bar Charts and Histograms**

Bar charts and histograms provide powerful tools for visualizing and comparing numerical data across different categories or distributions.

* **Bar Charts** are ideal for comparing numerical values across different categories:

````python
# Categories and values
categories = ['Category A', 'Category B', 'Category C']
values = [23, 45, 56]

# Creating a bar chart
plt.bar(categories, values)
plt.show()
````

* **Histograms** are used to show the distribution of a continuous data set, and they provide insights into the central tendency, variability, and shape of the data distribution.

````python
import numpy as np # to generate random numbers

# Generating a dataset
data = np.random.randn(1000)

# Creating a histogram
plt.hist(data, bins=15)
plt.show()
````

#### Visual Comparison: Bar Chart vs. Histogram

* **Bar Chart:** Useful for comparing different groups or categories. Each bar represents a category with a different value.
* **Histogram:** Best suited for showing the distribution of a dataset. It groups data into bins, showing the frequency of data points within each bin.

#### Choosing the Right Plot
* Use bar charts when comparing discrete categories or groups.
* Use histhograms to explore the distribution of continuous data sets, understand the central tendency, dispersion, and skewness.

#### Customization and Styling

Both bar charts and histograms can be customized in various ways in Matplotlib, including changing `colors` (a wide range of preset [colors](https://proplot.readthedocs.io/en/latest/colors.html) is available), adding `labels`, and adjusting the `bin` size (for histograms) to better represent the data.

````python
# Example: Customized Bar Chart
plt.bar(categories, values, color=['red', 'green', 'blue'])
plt.title('Colored Bar Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
````

````python
# Example: Customized Bar Chart

plt.hist(data, bins=20, color='skyblue', linestyle='--', linewidth=0.5, edgecolor='black')
plt.title('Customized Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
````

#### Styling Tips for Bar Charts and Histograms

* **Color Scheme:** Choose colors that are easy to distinguish from one another and are accessible to all viewers, including those with color vision deficiency.
* **Labels:** Always include clear labels for the x-axis, y-axis, and the chart itself. When applicable, adding labels to individual bars or segments can provide additional clarity.
* **Grid Lines:** Use grid lines sparingly. They can help in reading the chart, but too many can clutter the visualization.
* **Data Points:** Especially for histograms, showing the number of data points (n) can help contextualize the distribution.


````python
# Example: Customized Histogram

bars = plt.hist(data, bins=10)

# Adding value labels on top of each bar
n, bins, patches = plt.hist(data, bins=10)

# Adding value labels on top of each bar
for i in range(len(patches)):
    plt.text(patches[i].get_x() + patches[i].get_width() / 2, n[i] + 3, str(int(n[i])), ha='center', va='bottom')

plt.show()
````

By applying these customization and styling techniques, you can create bar charts and histograms that not only convey the essential data but do so in a visually appealing manner that engages your audience. Remember, the goal of visualization is not just to show numbers but to tell a story that those numbers represent.

#### **2.3 Scatter Plots**

Scatter plots are invaluable for visualizing the relationships between two variables, allowing us to observe patterns, trends, correlations, and even outliers within the data. They are particularly useful in identifying the type of relationship (linear, quadratic, etc.) that might exist between two variables.

Let's delve into creating a basic scatter plot, followed by more advanced customization options to enhance its readability and visual appeal.

````python
import numpy as np
import matplotlib.pyplot as plt

# Generating sample data
x = np.random.rand(50)
y = np.random.rand(50)

# Creating the scatter plot
plt.scatter(x, y)
plt.title('Simple Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
````

In this example, `x` and `y` are arrays of randomly generated numbers, and `plt.scatter(x, y)` creates a scatter plot of these points. The labels and title are added to make the plot informative.

#### Customizing Scatter Plots

Customization can significantly enhance the effectiveness of a scatter plot, making it easier to differentiate between data points or groups.

````python
#Example: Customized Scatter Plot

# Generating another set of sample data for comparison
x2 = np.random.rand(50) + 1  # Offset by 1 to distinguish
y2 = np.random.rand(50) + 1  # Offset by 1 to distinguish

# Creating a scatter plot with customized markers
plt.scatter(x, y, color='blue', marker='o', label='Group 1')
plt.scatter(x2, y2, color='red', marker='x', label='Group 2')

plt.title('Customized Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()  # Display legend to differentiate the groups
plt.show()
````

This customized example demonstrates how to plot two different groups in the same scatter plot, each with distinct marker styles and colors. Adding a legend helps in distinguishing these groups.

#### Enhancing Scatter Plots

Further enhancements can include adding grid lines for better readability, using transparency (alpha) to manage overlapping points, and sizing markers based on another variable to add a third dimension to the visualization.

````python
# Example: Advanced Scatter Plot

# Generating sample data with a third variable for size
size = np.random.rand(50) * 100  # Marker sizes

plt.scatter(x, y, s=size, color='blue', alpha=0.5, label='Data Points')
plt.title('Advanced Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.grid(True)
plt.show()
````

In this advanced example, the s parameter is used to vary the size of the markers based on another variable (`size` array), adding depth to the visualization. 

The `alpha` parameter adjusts the transparency, making it easier to visualize the density of overlapping points.

## **[3. Advanced Matplotlib Concepts](#4-advanced-visualization-techniques)**

Matplotlib's power lies in its vast array of customization options. This section explores how to fine-tune your plots, from adjusting plot dimensions to customizing plot styles for more engaging presentations.


### 3.1  Understanding Figures and Axes 

Before diving deeper into customization options, it's essential to grasp two core concepts in Matplotlib: `figures` and `axes`. 

* A **figure** in Matplotlib is the overall window or page that everything is drawn on. It's a top-level container for all plot elements. 

* The **axes** are the part of the figure where we plot our data and set labels, ticks, etc. A figure can contain multiple axes.

**Example 1:** Simple Usage of Figure and Axes

This simple example demonstrates how to create a figure and add a single set of axes to it, then plot a basic line graph.

````python
import matplotlib.pyplot as plt

# Create a new figure, with a default size
fig = plt.figure()

# Add an axes to the figure as part of the subplot arrangement 111 (1 row, 1 column, 1st subplot)
ax = fig.add_subplot(111)

# Plot data on the axes
ax.plot([0, 1, 2, 3], [10, 20, 25, 30], label='Example Data')
ax.set_title('Simple Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.legend()

plt.show()
````

**Example 2:** A More Complex Example with Multiple Axes

In this advanced example, we introduced a more detailed level of customization within a single axis. 

The use of a custom line style and marker enhances the visual appeal of the plot. This approach demonstrates the balance between simplicity and the capability for detailed customization that Matplotlib offers.

````python
import numpy as np

# Generate some data
x = np.linspace(0, 10, 100)
y = np.cos(x)

# Create a new figure
fig = plt.figure(figsize=(10, 6))

# Add an axes
ax = fig.add_subplot()

# Plot the data with a custom line style and marker
ax.plot(x, y, linestyle='--', color='r', marker='o', label='Cosine Wave')

# Customizations
ax.set_title('Advanced Customization Plot')
ax.set_xlabel('X Value')
ax.set_ylabel('Y Value')

# Adding a text annotation inside the plot
ax.text(5, 0, 'Midpoint', fontsize=12, ha='center')

# Adding a legend
ax.legend()

plt.show()
````

In **matplotlib**, the distinction between using `plt` (e.g. like using `plt.xlabel` to set a x-label) and `ax` (e.g. like using `ax.set_xlabel`)  arises from the two different interfaces provided by the library: the **MATLAB-style** *state-based* interface and the more *modern* **object-oriented (OO)** interface.

* **MATLAB-style State-based Interface**: This interface is designed to be similar to [MATLAB](https://de.mathworks.com/help/matlab/ref/plot.html)'s plotting interface. It keeps track of the current figure and axes, which allows you to plot and modify plots with simple commands, treating the state as global. For example, when you use `plt.xlabel()`, you're applying a label to the x-axis of the current axes in the current figure. This approach is straightforward and convenient for simple and interactive plotting, as it allows for quick modifications without needing to *explicitly* reference the figure or axes objects. 
    > It is conventient, but it can become less clear and manageable with more complex plots.

* **Object-oriented Interface**: The object-oriented interface is more flexible and recommended for *complex* figures (like those with multiple subplots). In this approach, you explicitly create figures and axes objects and call methods on them (like `ax.set_xlabel()` to set the x-axis label on the specific Axes object `ax`). This makes the code more readable and gives you finer control over the figure's properties, since you're working directly with the objects that compose the plot. 
    > It is more complex, but particularly useful in scripts and larger projects where clarity and maintainability are essential.

### 3.2 Working with Multiple Figures and Axes

Matplotlib allows for complex figure layouts, with multiple subplots each containing their own axes. This enables detailed comparisons and explorations of data.

#### **Simple Example:** Creating a 1 Row, 2 Columns Plot Layout

This example demonstrates how to create a simple layout with two plots side by side, focusing on plotting linear and quadratic functions. This straightforward approach is particularly useful for direct comparison of related datasets or different aspects of the same data.

````python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create figure and axes objects for a 1x2 layout
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot on the first axes
axs[0].plot(x, y1, 'tab:blue')
axs[0].set_title('Sine Wave')
axs[0].set_xlabel('X')
axs[0].set_ylabel('sin(X)')

# Plot on the second axes
axs[1].plot(x, y2, 'tab:orange')
axs[1].set_title('Cosine Wave')
axs[1].set_xlabel('X')
axs[1].set_ylabel('cos(X)')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
````

**Key Points:**

* **Simple Layout Creation:** The `plt.subplots(1, 2, figsize=(12, 4))` line initializes a figure with a `1x2` grid (one row, two columns) of subplots. This layout is ideal for showcasing a side-by-side comparison of two related plots.
* **Plotting and Customization:** Each subplot (linear and quadratic) is plotted with distinct colors for easy differentiation. Custom titles and axis labels succinctly convey the nature of each plot, enhancing the viewer's understanding of the data presented.
* **Adjusting Layout:** Utilizing `plt.tight_layout()` ensures that the plots and their labels are neatly organized within the figure without any overlap, optimizing the presentation for clarity and visual appeal.

#### **Intermediate Example:** Creating a 2 Rows, 2 Columns Plot Layout

In this intermediate example, we aim to demonstrate the versatility and control provided by Matplotlib's subplot feature through a `2x2` grid layout. This approach allows us to compare four distinct mathematical functions side by side, showcasing the library's capability to handle multiple plots within a single figure. 

Each subplot will display a different mathematical function: `Sine`, `Cosine`, `Tangent`, and `Hyperbolic Sine` waves, respectively. This not only highlights the use of subplots but also illustrates how to manipulate various aspects of each subplot for clarity and visual appeal.

````python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.sinh(x)

# Create figure and axes objects for a 2x2 layout
fig, axs = plt.subplots(2, 2, figsize=(10, 5))

# Flatten the axes array to make indexing easier
axs = axs.flatten()

# First plot: Sine Wave
axs[0].plot(x, y1, 'r-')
axs[0].set_title('Sine Wave')
axs[0].set_xlabel('X')
axs[0].set_ylabel('sin(X)')

# Second plot: Cosine Wave
axs[1].plot(x, y2, 'g-')
axs[1].set_title('Cosine Wave')
axs[1].set_xlabel('X')
axs[1].set_ylabel('cos(X)')

# Third plot: Tangent Wave
axs[2].plot(x, y3, 'b-')
axs[2].set_title('Tangent Wave')
axs[2].set_xlabel('X')
axs[2].set_ylabel('tan(X)')
axs[2].set_ylim([-5, 5])  # Limit y-axis to improve visibility

# Fourth plot: Hyperbolic Sine Wave
axs[3].plot(x, y4, 'm-')
axs[3].set_title('Hyperbolic Sine Wave')
axs[3].set_xlabel('X')
axs[3].set_ylabel('sinh(X)')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
````

**Key Points in This Example:**

* **Subplot Creation:** The `plt.subplots(2, 2, figsize=(10, 8))` call initializes a figure and creates a `2x2` grid for subplots. This setup is ideal for comparing different plots in a compact and organized manner.
* **Plotting Functions:** We plotted four different mathematical functions, demonstrating the ability to visualize a wide range of data within a single figure. This is crucial for comparative analysis and showcasing relationships between different datasets.
* **Customization:** Each subplot is individually customizable. Titles, axis labels, line colors, and styles are set explicitly for each subplot to enhance readability and aesthetic appeal.
* **Layout Adjustment:** The `plt.tight_layout()` call is used to automatically adjust the subplots' positions to prevent overlap and ensure that the figure looks organized and is easy to interpret.

This example effectively illustrates how to work with multiple subplots in Matplotlib, offering a balance between complexity and clarity, making it a valuable tool for data analysis and presentation.

#### **Complex Example:** Creating a 3 Rows, 3 Columns Plot Layout

In this more complex example, we'll expand our subplot arrangement into a 3x3 grid. This layout is excellent for in-depth comparative analysis across multiple datasets or variables, allowing for a comprehensive overview of different mathematical functions.

*Regarding `flatten`and `ravel`:* The primary functional difference is that `flatten` is a method of an ndarray object and hence can only be called for true numpy arrays. In contrast `ravel()` is a library-level function and hence can be called on any object that can successfully be parsed [(Source)](https://stackoverflow.com/questions/28837312/what-is-the-difference-between-flatten-and-ravel-in-numpy/28846614#28846614). 
* For example `ravel()` will work on a list of ndarrays, while `flatten` (obviously) won't.

````python
import matplotlib.pyplot as plt
import numpy as np

# Prepare the data: Generate x values and nine different y values for various functions
x = np.linspace(0, 2*np.pi, 100)
y_funcs = [np.sin(x), np.cos(x), np.tan(x), np.sinh(x), np.cosh(x), np.tanh(x), np.exp(x), np.log(x+1), np.sqrt(x)]

# Create a figure and a 3x3 grid of subplots
fig, axs = plt.subplots(3, 3, figsize=(10, 5))

# Plot each function in its respective subplot
for i, ax in enumerate(axs.ravel()):
    ax.plot(x, y_funcs[i])
    ax.set_title(f'Function {i+1}')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
````

**Key Insights:**

* **Complex Grid Setup:** The `plt.subplots(3, 3, figsize=(15, 15))` initialization creates a comprehensive `3x3` grid, enabling the detailed comparison of nine distinct plots within a single figure.
* **Diverse Data Visualization:** By showcasing a variety of functions, this example highlights Matplotlib's capability to handle and display a wide range of mathematical expressions, making it an invaluable tool for scientific and mathematical data visualization.
* **Efficient Plotting and Customization:** Utilizing a loop to plot each function streamlines the code and ensures consistency across plots. This method facilitates the efficient visualization of multiple datasets, while `plt.tight_layout()` optimizes the arrangement for maximum readability and aesthetic quality.

### [4.1 Pie charts](#41-pie-charts)

Pie charts are a type of data visualization that show the proportions of a dataset that belong to various categories as slices of a pie. Each slice's size is proportional to the quantity it represents. They are particularly useful for showing relative sizes and compositions within a whole, and they are most effective when comparing a limited number of categories to avoid clutter and ensure readability.

#### Creating a pie chart

In `matplotlib`, pie charts are created using the `pie` function of the pyplot module or the corresponding method of an Axes object in the object-oriented interface. 

Here's a simple [example](https://pypl.github.io/PYPL.html) of how to create a pie chart using `matplotlib`:

````python
# Data to plot
labels = ['Python', 'Java', 'JavaScript', 'C/C++', 'C#', 'R', 'Others']
sizes = [28.53, 16.04, 8.72, 6.65, 6.63, 4.63, 28.2]

# Plot
plt.pie(sizes,labels=labels)
  
plt.show()
````

````python
# Pie plot with some more stylings

# Data to plot
labels = ['Python', 'Java', 'JavaScript', 'C/C++', 'C#', 'R', 'Others']
sizes = [28.53, 16.04, 8.72, 6.65, 6.63, 4.63, 28.2]

colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6', '#d9d9d9']
explode = (0.1, 0, 0, 0, 0, 0, 0 )  # explode the 1st slice (Python)


plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)

plt.title("Popularity of programming languages")

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
````

These two examples demonstrated how to create a pie chart that represents the distribution of preferences for common programming languages. 

In the second example, the `explode` parameter *"explodes"* or offsets a slice from the pie (in this case, the 'Python' slice). Again, the `colors` parameter specifies the colors of each slice, and autopct adds labels to the slices with their percentage of the total.

### 4.3 Contour plots

Contour plots in `matplotlib` are a way to represent three-dimensional data on a two-dimensional plane. They are particularly useful for visualizing potential surfaces or elevation maps, where you want to see how a variable (z) changes with respect to two other variables (x and y) without having to create a 3D surface. Contour plots display lines or curves (contours) that connect points of equal value.

#### Creating a simple contour plot

To create a contour plot, you generally need a grid of `x`, `y` values and a corresponding matrix of `z` values. You can use the `np.meshgrid` function to create the `x` and `y` grids, and then compute `z` for each (`x`, `y`) pair.

````python
import matplotlib.pyplot as plt
import numpy as np

# Generate grid points
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
x, y = np.meshgrid(x, y)
z = np.sin(x) * np.cos(y)

# Create a contour plot
plt.figure()
contour_set = plt.contour(x, y, z, cmap='viridis')
plt.clabel(contour_set, inline=True, fontsize=8)
plt.title('Simple Contour Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
````

## 4.4 3D Plots

3D plotting in matplotlib is an extension of the usual 2D plotting capabilities, allowing for the visualization of three-dimensional data. This is achieved through the mplot3d toolkit, which adds support for simple 3D graphs like scatter plots, surface plots, wireframes, and contour plots. 3D plots are particularly useful for exploring complex datasets where relationships between three variables need to be visualized simultaneously.

#### Creating a 3D Surface Plot

To create 3D plots, you first need to import the `Axes3D` object from the `mpl_toolkits.mplot3d` module. This object creates a 3D axis to be used for the plot.

````python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Plot a 3D surface
surface = ax.plot_surface(x, y, z, cmap='plasma')
fig.colorbar(surface)

plt.title('3D Surface Plot')
plt.show()
````

### 4.5 And many more...

There is a wide range of other plots available in `matplotlib` as well: https://matplotlib.org/stable/plot_types/index.html

The example gallery is a great starting point if you want to explore more options for plotting your data: https://matplotlib.org/stable/gallery/index.html

Here is one final example of what else you can do with `matplotlib`:

````python
from matplotlib.tri import Triangulation

# Generate data for each plot type
# Triplot data
x_tri = np.random.rand(100, 2)
tri = Triangulation(x_tri[:, 0], x_tri[:, 1])

# Barbs data
x_barbs = np.linspace(-2, 2, 10)
y_barbs = np.linspace(-2, 2, 10)
u, v = np.meshgrid(x_barbs, y_barbs)

# Streamplot data
Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2

# Tripcolor data
x_tripcolor = np.random.rand(100, 2)
tri_tripcolor = Triangulation(x_tripcolor[:, 0], x_tripcolor[:, 1])
z_tripcolor = np.sin(x_tripcolor[:, 0]) + np.cos(x_tripcolor[:, 1])

# Hexbin data
x_hexbin = np.random.randn(1000)
y_hexbin = np.random.randn(1000)

# Eventplot data
data_eventplot = np.random.random([5, 100]) * 100

# Boxplot data
data_boxplot = np.random.normal(size=(100, 3))

# Errorbar data
x_errorbar = np.arange(0.1, 4, 0.5)
y_errorbar = np.exp(-x_errorbar)
errorbar_lower = 0.1 * np.sqrt(x_errorbar)
errorbar_upper = 0.2 * np.sqrt(x_errorbar)

# Violinplot data
data_violinplot = np.random.normal(size=(100, 3))

# Creating the 3x3 plot
fig, axs = plt.subplots(3, 3, figsize=(10, 8))

# Triplot
axs[0, 0].triplot(tri, 'bo-')
axs[0, 0].set_title('Triplot')

# Barbs
axs[0, 1].barbs(x_barbs, y_barbs, u, v)
axs[0, 1].set_title('Barbs')

# Streamplot
axs[0, 2].streamplot(X, Y, U, V)
axs[0, 2].set_title('Streamplot')

# Tripcolor
axs[1, 0].tripcolor(tri_tripcolor, z_tripcolor, shading='flat')
axs[1, 0].set_title('Tripcolor')

# Hexbin
axs[1, 1].hexbin(x_hexbin, y_hexbin, gridsize=30, cmap='Blues')
axs[1, 1].set_title('Hexbin')

# Eventplot
axs[1, 2].eventplot(data_eventplot)
axs[1, 2].set_title('Eventplot')

# Boxplot
axs[2, 0].boxplot(data_boxplot)
axs[2, 0].set_title('Boxplot')

# Errorbar
axs[2, 1].errorbar(x_errorbar, y_errorbar, yerr=[errorbar_lower, errorbar_upper], fmt='o')
axs[2, 1].set_title('Errorbar')

# Violinplot
axs[2, 2].violinplot(data_violinplot)
axs[2, 2].set_title('Violinplot')

plt.tight_layout()
plt.show()
````
