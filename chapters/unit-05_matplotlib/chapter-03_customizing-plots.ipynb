## **Chapter 3:** Advanced Matplotlib Concepts

Matplotlib's power lies in its vast array of customization options. This section explores how to fine-tune your plots, from adjusting plot dimensions to customizing plot styles for more engaging presentations.


### 3.1  Understanding Figures and Axes 

Before diving deeper into customization options, it's essential to grasp two core concepts in Matplotlib: `figures` and `axes`. 

* A **figure** in Matplotlib is the overall window or page that everything is drawn on. It's a top-level container for all plot elements. 

* The **axes** are the part of the figure where we plot our data and set labels, ticks, etc. A figure can contain multiple axes.

**Example 1:** Simple Usage of Figure and Axes

This simple example demonstrates how to create a figure and add a single set of axes to it, then plot a basic line graph.

```python
import matplotlib.pyplot as plt

# Create a new figure, with a default size
fig = plt.figure()

# Add an axes to the figure as part of the subplot arrangement 111 (1 row, 1 column, 1st subplot)
ax = fig.add_subplot(111)

# Plot data on the axes
ax.plot([0, 1, 2, 3], [10, 20, 25, 30], label='Example Data')
ax.set_title('Simple Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.legend()

plt.show()
```

**Example 2:** A More Complex Example with Multiple Axes

In this advanced example, we introduced a more detailed level of customization within a single axis. 

The use of a custom line style and marker enhances the visual appeal of the plot. This approach demonstrates the balance between simplicity and the capability for detailed customization that Matplotlib offers.

```python
import numpy as np

# Generate some data
x = np.linspace(0, 10, 100)
y = np.cos(x)

# Create a new figure
fig = plt.figure(figsize=(10, 6))

# Add an axes
ax = fig.add_subplot()

# Plot the data with a custom line style and marker
ax.plot(x, y, linestyle='--', color='r', marker='o', label='Cosine Wave')

# Customizations
ax.set_title('Advanced Customization Plot')
ax.set_xlabel('X Value')
ax.set_ylabel('Y Value')

# Adding a text annotation inside the plot
ax.text(5, 0, 'Midpoint', fontsize=12, ha='center')

# Adding a legend
ax.legend()

plt.show()
```

In **matplotlib**, the distinction between using `plt` (e.g. like using `plt.xlabel` to set a x-label) and `ax` (e.g. like using `ax.set_xlabel`)  arises from the two different interfaces provided by the library: the **MATLAB-style** *state-based* interface and the more *modern* **object-oriented (OO)** interface.

* **MATLAB-style State-based Interface**: This interface is designed to be similar to [MATLAB](https://de.mathworks.com/help/matlab/ref/plot.html)'s plotting interface. It keeps track of the current figure and axes, which allows you to plot and modify plots with simple commands, treating the state as global. For example, when you use `plt.xlabel()`, you're applying a label to the x-axis of the current axes in the current figure. This approach is straightforward and convenient for simple and interactive plotting, as it allows for quick modifications without needing to *explicitly* reference the figure or axes objects. 
    > It is conventient, but it can become less clear and manageable with more complex plots.

* **Object-oriented Interface**: The object-oriented interface is more flexible and recommended for *complex* figures (like those with multiple subplots). In this approach, you explicitly create figures and axes objects and call methods on them (like `ax.set_xlabel()` to set the x-axis label on the specific Axes object `ax`). This makes the code more readable and gives you finer control over the figure's properties, since you're working directly with the objects that compose the plot. 
    > It is more complex, but particularly useful in scripts and larger projects where clarity and maintainability are essential.

### 3.2 Working with Multiple Figures and Axes

Matplotlib allows for complex figure layouts, with multiple subplots each containing their own axes. This enables detailed comparisons and explorations of data.

#### **Simple Example:** Creating a 1 Row, 2 Columns Plot Layout

This example demonstrates how to create a simple layout with two plots side by side, focusing on plotting linear and quadratic functions. This straightforward approach is particularly useful for direct comparison of related datasets or different aspects of the same data.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create figure and axes objects for a 1x2 layout
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot on the first axes
axs[0].plot(x, y1, 'tab:blue')
axs[0].set_title('Sine Wave')
axs[0].set_xlabel('X')
axs[0].set_ylabel('sin(X)')

# Plot on the second axes
axs[1].plot(x, y2, 'tab:orange')
axs[1].set_title('Cosine Wave')
axs[1].set_xlabel('X')
axs[1].set_ylabel('cos(X)')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

**Key Points:**

* **Simple Layout Creation:** The `plt.subplots(1, 2, figsize=(12, 4))` line initializes a figure with a `1x2` grid (one row, two columns) of subplots. This layout is ideal for showcasing a side-by-side comparison of two related plots.
* **Plotting and Customization:** Each subplot (linear and quadratic) is plotted with distinct colors for easy differentiation. Custom titles and axis labels succinctly convey the nature of each plot, enhancing the viewer's understanding of the data presented.
* **Adjusting Layout:** Utilizing `plt.tight_layout()` ensures that the plots and their labels are neatly organized within the figure without any overlap, optimizing the presentation for clarity and visual appeal.

#### **Intermediate Example:** Creating a 2 Rows, 2 Columns Plot Layout

In this intermediate example, we aim to demonstrate the versatility and control provided by Matplotlib's subplot feature through a `2x2` grid layout. This approach allows us to compare four distinct mathematical functions side by side, showcasing the library's capability to handle multiple plots within a single figure. 

Each subplot will display a different mathematical function: `Sine`, `Cosine`, `Tangent`, and `Hyperbolic Sine` waves, respectively. This not only highlights the use of subplots but also illustrates how to manipulate various aspects of each subplot for clarity and visual appeal.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.sinh(x)

# Create figure and axes objects for a 2x2 layout
fig, axs = plt.subplots(2, 2, figsize=(10, 5))

# Flatten the axes array to make indexing easier
axs = axs.flatten()

# First plot: Sine Wave
axs[0].plot(x, y1, 'r-')
axs[0].set_title('Sine Wave')
axs[0].set_xlabel('X')
axs[0].set_ylabel('sin(X)')

# Second plot: Cosine Wave
axs[1].plot(x, y2, 'g-')
axs[1].set_title('Cosine Wave')
axs[1].set_xlabel('X')
axs[1].set_ylabel('cos(X)')

# Third plot: Tangent Wave
axs[2].plot(x, y3, 'b-')
axs[2].set_title('Tangent Wave')
axs[2].set_xlabel('X')
axs[2].set_ylabel('tan(X)')
axs[2].set_ylim([-5, 5])  # Limit y-axis to improve visibility

# Fourth plot: Hyperbolic Sine Wave
axs[3].plot(x, y4, 'm-')
axs[3].set_title('Hyperbolic Sine Wave')
axs[3].set_xlabel('X')
axs[3].set_ylabel('sinh(X)')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

**Key Points in This Example:**

* **Subplot Creation:** The `plt.subplots(2, 2, figsize=(10, 8))` call initializes a figure and creates a `2x2` grid for subplots. This setup is ideal for comparing different plots in a compact and organized manner.
* **Plotting Functions:** We plotted four different mathematical functions, demonstrating the ability to visualize a wide range of data within a single figure. This is crucial for comparative analysis and showcasing relationships between different datasets.
* **Customization:** Each subplot is individually customizable. Titles, axis labels, line colors, and styles are set explicitly for each subplot to enhance readability and aesthetic appeal.
* **Layout Adjustment:** The `plt.tight_layout()` call is used to automatically adjust the subplots' positions to prevent overlap and ensure that the figure looks organized and is easy to interpret.

This example effectively illustrates how to work with multiple subplots in Matplotlib, offering a balance between complexity and clarity, making it a valuable tool for data analysis and presentation.

#### **Complex Example:** Creating a 3 Rows, 3 Columns Plot Layout

In this more complex example, we'll expand our subplot arrangement into a 3x3 grid. This layout is excellent for in-depth comparative analysis across multiple datasets or variables, allowing for a comprehensive overview of different mathematical functions.

*Regarding `flatten`and `ravel`:* The primary functional difference is that `flatten` is a method of an ndarray object and hence can only be called for true numpy arrays. In contrast `ravel()` is a library-level function and hence can be called on any object that can successfully be parsed [(Source)](https://stackoverflow.com/questions/28837312/what-is-the-difference-between-flatten-and-ravel-in-numpy/28846614#28846614). 
* For example `ravel()` will work on a list of ndarrays, while `flatten` (obviously) won't.

```python
import matplotlib.pyplot as plt
import numpy as np

# Prepare the data: Generate x values and nine different y values for various functions
x = np.linspace(0, 2*np.pi, 100)
y_funcs = [np.sin(x), np.cos(x), np.tan(x), np.sinh(x), np.cosh(x), np.tanh(x), np.exp(x), np.log(x+1), np.sqrt(x)]

# Create a figure and a 3x3 grid of subplots
fig, axs = plt.subplots(3, 3, figsize=(10, 5))

# Plot each function in its respective subplot
for i, ax in enumerate(axs.ravel()):
    ax.plot(x, y_funcs[i])
    ax.set_title(f'Function {i+1}')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

**Key Insights:**

* **Complex Grid Setup:** The `plt.subplots(3, 3, figsize=(15, 15))` initialization creates a comprehensive `3x3` grid, enabling the detailed comparison of nine distinct plots within a single figure.
* **Diverse Data Visualization:** By showcasing a variety of functions, this example highlights Matplotlib's capability to handle and display a wide range of mathematical expressions, making it an invaluable tool for scientific and mathematical data visualization.
* **Efficient Plotting and Customization:** Utilizing a loop to plot each function streamlines the code and ensures consistency across plots. This method facilitates the efficient visualization of multiple datasets, while `plt.tight_layout()` optimizes the arrangement for maximum readability and aesthetic quality.

### 3.3 Custom Annotations and Styles

Enhance the readability and attractiveness of your plots by adding annotations, using different line styles and markers, and applying custom themes.

#### Custom Annotations

Annotations can be used to highlight specific points in your data or to add text explanations directly on the plot. They can be particularly useful for identifying outliers, marking special events, or adding extra context to your visualizations.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.linspace(0, 10, 100)
y = np.cos(x)

plt.figure()
plt.plot(x, y)

# Highlight a specific point with an annotation
plt.annotate('Local Minimum', xy=(np.pi, -1), xytext=(np.pi+1, -1.5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             )

plt.title('Plot with Custom Annotation')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

This example demonstrates how to use plt.annotate() to add a text label and an arrow pointing to a specific point on the plot, enhancing the plot's informativeness.

#### Using Styles

* Matplotlib's **style** module allows you to use [predefined styles](https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html). This can greatly enhance the aesthetics of your visualizations with minimal code (learn [more](https://matplotlib.org/stable/users/explain/customizing.html#customizing-with-style-sheets)). 

You can apply predefined styles or customize your own to ensure your plots match your presentation or publication's aesthetic requirements.

#### Applying a Predefined Style

```python
# Example: Using a Predefined Style

# Setting the style
plt.style.use('dark_background') # try: "ggplot", "grayscale", "classic"

# Creating two line plots
plt.plot(x, y)
plt.plot(x, x)
  
plt.show()

# Run this line so that ipython resets the plot style
plt.style.use('default')
```

In this example, we use `plt.style.use()` to apply the `dark_background` style, which is one of the many available predefined styles. 

This changes the plot's background, gridlines, and color palette to fit the chosen style, making it more visually appealing with minimal effort.

**Step 3:** Create a figure to display twelve charts in a (4x3 grid) 

* Use all measurements (except for the `time_to_fill` and the `quality`)

* Use annotations for the charts and use different [colors](https://matplotlib.org/3.1.0/tutorials/colors/colors.html) to distinguish them.

* Finally, [pick](https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html) a custom `style` for your plot. 


```python
# Create a figure to display twelve charts in a (4x3 grid) 

# Setting the style
plt.style.use('dark_background') # try: "ggplot", "grayscale", "classic"

colors = [
    '#100C08', # hex
    '#738678',
    '#123456',
    (1,0,0), # rgb
    (1,0.5,0),
    (1,0,0.5),
    'tab:blue', # tableau colors
    'tab:green',
    'tab:red',
    'c', # base colors
    'b',
    'y'
]

default_font_size = 8

# Create a figure and a 3x3 grid of subplots
fig, axes = plt.subplots(4, 3, figsize=(10, 8))

# Iterate axes and column names 
for i, (ax, colum_name) in enumerate(zip(axes.ravel(), df_data.columns)):
    # Plot
    ax.plot(df_data[colum_name], label=colum_name, color=colors[i])
    # Annotations
    ax.set_title(colum_name, fontsize=default_font_size)
    ax.set_xlabel("Number of recordings", fontsize=default_font_size)
    ax.set_ylabel("Measurement", fontsize=default_font_size)
    # Change tick font size
    ax.xaxis.set_tick_params(labelsize=default_font_size-2)
    ax.yaxis.set_tick_params(labelsize=default_font_size-2)


plt.suptitle("Visualization of the IM process data")

plt.tight_layout()
plt.show()

plt.style.use('default')
```
